```python
th = soup.find_all('th')
td = soup.find_all('td')
td
```

```python
t = [0,1,2]
```

```python
Country = [] 
for i in range(1,len(td),3):
    Country.append(td[i].text)
Life = []

for i in range(2,len(td),3):
    Life.append(td[i])

print(Country)
print(Life)
```

```python
Life = Life[:43]
```
